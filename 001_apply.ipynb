```python
import pandas as pd
```

```python
pokemons = pd.read_csv("Pokemon.csv", index_col="#")
pokemons
```

```python
pokemons["name_length"] = pokemons["Name"].apply(len)
pokemons
```

```python
pokemons["name_length"]
```

```python
def strength(attack:int) -> str:
    if attack > 150:
        return "Stromg"
    elif attack > 100:
        return "Average"
    else:
        return "Week"
```

```python
pokemons["Strength"] = pokemons["Attack"].apply(strength)
pokemons
```

```python
del pokemons["name_length"]
```

```python
def score(row):
    attack: int = row["Attack"]
    defence: int = row["Defense"]
    speed: int = row["Speed"]
    return sum([attack, defence, speed])
```

```python
pokemons["score"] = pokemons.apply(score, axis=1)
pokemons
```

```python
short_name = pokemons["Name"].apply(lambda name: name[:5])
short_name
```
