```python
import pandas as pd
```

```python
cols = ["name", "survived", "pclass", "fare", "age", "sex"]
titanic = pd.read_excel("titanic3.xls", usecols=cols, index_col="name")
titanic
```

```python
uper_age_mask = titanic["age"] >= 20
lower_age_mask = titanic["age"] <= 25
```

```python
titanic[uper_age_mask & lower_age_mask]
```

```python
youth_mask = titanic["age"].between(20, 25)
titanic[youth_mask]
```

```python
youth_mask = titanic["age"].between(20, 25, inclusive='neither')
titanic[youth_mask]
```
